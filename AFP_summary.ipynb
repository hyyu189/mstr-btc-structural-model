This notebook summarizes the calibrated parameters, indicators, and core plots produced by the structural model pipeline.

This will regenerate the processed panel, calibrated parameters, simulation results, and figures under the `results/` directory.


````python
from pathlib import Path
import json

import pandas as pd
from IPython.display import Image, display

RESULTS_DIR = Path("results")
FIGURES_DIR = RESULTS_DIR / "figures"

# Load processed daily panel
panel = pd.read_csv(RESULTS_DIR / "panel_processed.csv", parse_dates=["date"]).set_index("date")

# Load calibrated parameters and indicators
with open(RESULTS_DIR / "params.json", "r", encoding="utf-8") as f:
    params = json.load(f)
with open(RESULTS_DIR / "indicators.json", "r", encoding="utf-8") as f:
    indicators = json.load(f)

panel.shape, panel.index.min(), panel.index.max()
````

## Panel Overview

The panel contains daily BTC price, MSTR holdings, debt, equity, premium, and BTC per share, aligned from the first Compustat daily observation through the target end date.

## Core Plots

The following figures are generated by the pipeline and saved under `results/figures/`:

- Assets, debt, and NAV over time.
- Premium time series.
- BTC per share time series.
- ILE and TEE over time.
- IFRD histogram from 3-year Monte Carlo simulations.


````python
figure_files = [
    "assets_debt_nav.png",
    "premium_timeseries.png",
    "btc_per_share_timeseries.png",
    "ile_tee_timeseries.png",
    "ifrd_histogram.png",
]

for name in figure_files:
    path = FIGURES_DIR / name
    if path.exists():
        print(name)
        display(Image(filename=str(path)))
    else:
        print(f"Missing figure: {path}")
````
